```python
import pandas as pd
import numpy as np
import os
import string
```

```python
data_cleaned = pd.read_csv('/content/drive/MyDrive/data_cleaned_ty1.csv')
```

```python
data_stemmed = pd.read_csv('/content/drive/MyDrive/data_stemmed_ty1.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = data_cleaned['recommendation'].astype('category').cat.codes

X_train, X_test, y_train, y_test = train_test_split(list(data_cleaned['reviews']), y, 
                                                   test_size=0.2, 
                                                   random_state = 0, 
                                                   #stratify = y,
                                                   shuffle=True)
```

```python
max_length = 500
```

```python
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization

vectorizer = TextVectorization(max_tokens=20000, output_sequence_length=max_length)
vectorizer.adapt(X_train)
```

```python
vectorizer.get_vocabulary()[:6]
```

```python
import os 

path_to_glove_file = os.path.join(
    os.path.expanduser("~"), "/content/drive/MyDrive/vocab.txt"
)

embeddings_index = {}
with open(path_to_glove_file) as f:
    for line in f:
        word, coefs = line.split(maxsplit=1)
        coefs = np.fromstring(coefs, "f", sep=" ")
        embeddings_index[word] = coefs

print("Found %s word vectors." % len(embeddings_index))
```

```python
# Préparation de la matrix d'embedding

voc = vectorizer.get_vocabulary()
word_index = dict(zip(voc, range(len(voc))))

num_tokens = len(voc) + 2
embedding_dim = 300
hits = 0
misses = 0


embedding_matrix = np.zeros((num_tokens, embedding_dim))
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
        hits += 1
    else:
        misses += 1
print("Converted %d words (%d misses)" % (hits, misses))
```
